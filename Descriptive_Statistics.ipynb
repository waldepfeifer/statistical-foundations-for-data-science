# Data Analysis of Adult School Dataset

Here you have a dataset ('Students.csv') from an adult school that offers education to young adults (around 24 years old). The school has recently tested a new learning method. The collected variables are: 
- Gender (Male/Female)
- SocioEconomicStatus (Low/Medium/High)
- Age (Numerical)
- LearningMethod (Old/New)
- InitialGrade (Numerical)
- FinalGrade (Numerical)
The school made a test to check the level of each students (InitalGrade) before starting the course. Then, each student was assigned to a specific learning method (Old or New). After the course, the students took again another test to verify their final level (FinalGrade).

Use different descriptive statistical techniques to analyze the data.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 500)
```

```python
df = pd.read_csv('students.csv')
df
```

## Exercise 1
Create the table of absolute and relative frequencies for the variables Gender, Socio-economic Status, and Method of Learning.

```python
#Answer to Exercise 1: Gender
gender_freq = df['Gender'].value_counts().reset_index()
gender_freq.columns = ['Gender', 'Absolute Frequency']

# Calculating relative frequency
gender_freq['Relative Frequency'] = gender_freq['Absolute Frequency'] / len(df)

gender_freq
```

```python
#Answer to Exercise 1: SocioEconomicStatus
socio_freq = df['SocioEconomicStatus'].value_counts().reset_index()
socio_freq.columns = ['SocioEconomicStatus', 'Absolute Frequency']

# Calculating relative frequency
socio_freq['Relative Frequency'] = socio_freq['Absolute Frequency'] / len(df)

socio_freq
```

```python
#Answer to Exercise 1: LearningMethod
learn_freq = df['LearningMethod'].value_counts().reset_index()
learn_freq.columns = ['LearningMethod', 'Absolute Frequency']

# Calculating relative frequency
learn_freq['Relative Frequency'] = learn_freq['Absolute Frequency'] / len(df)

learn_freq
```

## Exercise 2
Create pie charts and bar charts to describe the previous variables.

```python
# Pie chart for gender distribution (Relative)
plt.figure(figsize=(3,3))
plt.pie(gender_freq['Relative Frequency'], labels=gender_freq['Gender'], autopct='%1.1f%%', startangle=90)
plt.title('Gender Distribution (Pie Chart - Relative Frequency)')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.tight_layout()
plt.show()


# Bar chart for gender distribution (Absolute)
plt.figure(figsize=(4,2))
plt.bar(gender_freq['Gender'], gender_freq['Absolute Frequency'], color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Absolute Frequency')
plt.title('Gender Distribution (Bar Chart)')
plt.tight_layout()
plt.show()
```

```python
# Pie chart for SocioEconomicStatus distribution (Relative)
plt.figure(figsize=(3,3))
plt.pie(socio_freq['Relative Frequency'], labels=socio_freq['SocioEconomicStatus'], autopct='%1.1f%%', startangle=90)
plt.title('SocioEconomicStatus Distribution (Pie Chart - Relative Frequency)')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.tight_layout()
plt.show()


# Bar chart for SocioEconomicStatus distribution (Absolute)
plt.figure(figsize=(4,2))
plt.bar(socio_freq['SocioEconomicStatus'], socio_freq['Absolute Frequency'], color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Absolute Frequency')
plt.title('SocioEconomicStatus Distribution (Bar Chart)')
plt.tight_layout()
plt.show()
```

```python
# Pie chart for LearningMethod distribution (Relative)
plt.figure(figsize=(3,3))
plt.pie(learn_freq['Relative Frequency'], labels=learn_freq['LearningMethod'], autopct='%1.1f%%', startangle=90)
plt.title('LearningMethod Distribution (Pie Chart - Relative Frequency)')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.tight_layout()
plt.show()


# Bar chart for LearningMethod distribution (Absolute)
plt.figure(figsize=(4,2))
plt.bar(learn_freq['LearningMethod'], learn_freq['Absolute Frequency'], color='skyblue')
plt.xlabel('Gender')
plt.ylabel('Absolute Frequency')
plt.title('LearningMethod Distribution (Bar Chart)')
plt.tight_layout()
plt.show()
```

## Exercise 3
Create a histogram with the variable Age. Calculate the descriptive statistics: mean, median, mode, variance, standard deviation, quartiles, and IQR. Also, create the density plot and the boxplot.

```python
#Answer to Exercise 3

# Histogram for Age
plt.figure(figsize=(8,6))
plt.hist(df['Age'], bins=15, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.tight_layout()
plt.show()

# Density plot for Age
plt.figure(figsize=(8,6))
sns.kdeplot(df['Age'], fill=True, color='purple')
plt.title('Density Plot of Age')
plt.tight_layout()
plt.show()

# Boxplot for Age
plt.figure(figsize=(8,6))
sns.boxplot(x=df['Age'], color='lightgreen')
plt.title('Boxplot of Age')
plt.tight_layout()
plt.show()

# Descriptive statistics
age_mean = df['Age'].mean()
age_median = df['Age'].median()
age_mode = df['Age'].mode()[0]
age_variance = df['Age'].var()
age_std_dev = df['Age'].std()
age_quartiles = df['Age'].quantile([0.25, 0.5, 0.75])
age_iqr = age_quartiles[0.75] - age_quartiles[0.25]

# Displaying the descriptive statistics
print(f"Mean: {age_mean}")
print(f"Median: {age_median}")
print(f"Mode: {age_mode}")
print(f"Variance: {age_variance}")
print(f"Standard Deviation: {age_std_dev}")
print(f"Quartiles: {age_quartiles}")
print(f"IQR: {age_iqr}")
```

## Exercise 6
Draw two bar charts side by side. The first chart should split the socio-economic status by gender. And the second graph should split the gender by socio-economic status. Repeat the bar charts with stacked charts and percentages.


```python
# Define the order for Socio-Economic Status
socio_economic_order = ['Low', 'Medium', 'High']

# Set up the figure for two side-by-side bar charts
fig, axes = plt.subplots(1, 2, figsize=(14, 6), dpi=100)

# First bar chart: Socio-Economic Status by Gender (Regular Bar Chart)
sns.countplot(x='SocioEconomicStatus', hue='Gender', data=df, ax=axes[0], palette='Blues', order=socio_economic_order)
axes[0].set_title('Socio-Economic Status by Gender')
axes[0].set_xlabel('Socio-Economic Status')
axes[0].set_ylabel('Count')

# Second bar chart: Gender by Socio-Economic Status (Regular Bar Chart)
sns.countplot(x='Gender', hue='SocioEconomicStatus', data=df, ax=axes[1], palette='Blues', hue_order=socio_economic_order)
axes[1].set_title('Gender by Socio-Economic Status')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')

# Add space between the two charts
plt.subplots_adjust(wspace=0.3)
plt.tight_layout()
plt.show()

# Stacked Bar Charts
fig, axes = plt.subplots(1, 2, figsize=(14, 6), dpi=100)

# First stacked bar chart: Socio-Economic Status by Gender (Stacked)
gender_socioeconomic_counts = pd.crosstab(df['SocioEconomicStatus'], df['Gender'])
gender_socioeconomic_counts = gender_socioeconomic_counts.reindex(socio_economic_order)  # Ensure the index is sorted by socio-economic status
gender_socioeconomic_counts.plot(kind='bar', stacked=True, figsize=(6, 6), color=['lightblue', 'blue'], ax=axes[0])
axes[0].set_title('Status by Gender (Stacked)')
axes[0].set_xlabel('Socio-Economic Status')
axes[0].set_ylabel('Count')

# Second stacked bar chart: Gender by Socio-Economic Status (Stacked)
socioeconomic_gender_counts = pd.crosstab(df['Gender'], df['SocioEconomicStatus'])
socioeconomic_gender_counts = socioeconomic_gender_counts[socio_economic_order]  # Ensure the columns are sorted by socio-economic status
socioeconomic_gender_counts.plot(kind='bar', stacked=True, figsize=(6, 6), color=['lightblue', 'blue', 'darkblue'], ax=axes[1])
axes[1].set_title('Gender by Status (Stacked)')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Count')

# Add space between the two charts
plt.subplots_adjust(wspace=0.3)
plt.tight_layout()
plt.show()

# Percentage Bar Charts
fig, axes = plt.subplots(1, 2, figsize=(14, 6), dpi=100)

# First percentage bar chart: Socio-Economic Status by Gender (Percentage)
gender_socioeconomic_percentage = gender_socioeconomic_counts.div(gender_socioeconomic_counts.sum(axis=1), axis=0) * 100
gender_socioeconomic_percentage.plot(kind='bar', stacked=True, figsize=(6, 6), color=['lightblue', 'blue'], ax=axes[0])
axes[0].set_title('Status by Gender (Percentage)')
axes[0].set_xlabel('Socio-Economic Status')
axes[0].set_ylabel('Percentage')

# Second percentage bar chart: Gender by Socio-Economic Status (Percentage)
socioeconomic_gender_percentage = socioeconomic_gender_counts.div(socioeconomic_gender_counts.sum(axis=1), axis=0) * 100
socioeconomic_gender_percentage.plot(kind='bar', stacked=True, figsize=(6, 6), color=['lightblue', 'blue', 'darkblue'], ax=axes[1])
axes[1].set_title('Gender by Status (Percentage)')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Percentage')

plt.tight_layout()
plt.show()
```

## Exercise 7
Compute the mean and the standard deviation of Age split by Gender. Draw the boxplots side by side.

Draw the scatter plot between variables Initial Grades and Final Grades. Calculate the correlation coefficient.
